```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
display(df.describe())
```

```python
df.Outcome.value_counts()
```

```python
x = df.drop('Outcome', axis=1)
y = df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,stratify=y)
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.ensemble import BaggingClassifier

scores = cross_val_score(DecisionTreeClassifier(),x,y,cv=10) # cv=10 -> total datasets divided into 10 mini datasets
print('before ensemble \navg: ', scores.mean())

bag_model = BaggingClassifier(
    estimator=LogisticRegression(max_iter=1000),
    n_estimators=1000, # decides the number of sub-groups of dataset and their correspinding models
    max_samples=0.8, # chooses from only 80% of the total data
    oob_score=True, # gives the accuracy of the model tested on data that were not selected from the train data(80% that we have selected as max_samples parameter) in the subgroups of data
    random_state=0
)
bag_model.fit(x_train, y_train)
print('after ensemble')
print('oob  : ', bag_model.oob_score_)
print('train: ', bag_model.score(x_train,y_train))
print('test : ', bag_model.score(x_test,y_test))
```
